# Effective Area

This example will produce the effective areas for a set of XRT x-ray channels paired with thicknesses of the CCD contamination layer at a given date.

```python
import xrtpy
```

Define a filter and date

```python
Filter = "Al_Poly"
Date ="2006-09-22T22:45:45"
```

Reference the Effective Area Fundamental. This is where you'll find all functions & properties used for calculate the effective area.

```python
Effective_Area_Fundamental = xrtpy.response.EffectiveAreaFundamental(Filter, Date)
```

Define the effective area function 

```python
effective_area = Effective_Area_Fundamental.effective_area()
```

Define the wavelength 

```python
wavelength = Effective_Area_Fundamental.channel_wavelength
```

Create a plotting function that plots the effective area versus wavelegth. 

```python
def plot_effective_area():
    import matplotlib.pyplot as plt
    plt.figure(figsize=(30, 13))

    plt.plot(wavelength, effective_area,label='{:}'.format(Filter))
    
    plt.title('X-Ray Telescope \nObservation Date {:}'.format(Date),fontsize = 30)
    plt.xlabel('Wavelength (Ã…)',fontsize = 30)
    plt.ylabel('Effective Area $cm^{2}$',fontsize=30)
    plt.legend(fontsize=30)
    
    plt.xticks(fontsize=27)
    plt.yticks(fontsize=27)
    plt.xlim(0,60)

    plt.grid(color='lightgrey')
    plt.show()  
```

Run plot effective area function

```python
plot_effective_area()
```
