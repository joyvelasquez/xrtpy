# Channel properties

Channel explores some properties about the X-Ray Telescope (XRT) instrument. 

```python
import xrtpy
```

Begin by creating a Channel object by selecting a filter of interest. For this example we'll use Ti-poly. The process is the same for all xrt filters. 

```python
channel =  xrtpy.response.Channel("Ti-poly")
```

We can print basic information about XRT. 

```python
print("Choosen filter:",channel.name)
print("\nObservatory:",channel.observatory)
print("Instrument:",channel.instrument)
```

#### XRT Geometry

```python
print(channel.geometry.name,'\n')
print('Geometry focal len:',channel.geometry.focal_len)
print('Geometry aperture area:',channel.geometry.aperture_area)
```

#### XRT charge-coupled devices (CCD)

```python
print(channel.ccd.ccd_name)

print('\nCCD eV pre electron: ',channel.ccd.ccd_ev_ore_electron)
print('CCD full well: ',channel.ccd.ccd_full_well)
print('CCD gain left: ',channel.ccd.ccd_gain_left)
print('CCD gain right: ',channel.ccd.ccd_gain_right)
print('CCD pixel size: ',channel.ccd.ccd_pixel_size)
```

#### Plotting XRT properties - Transmittance

Define XRT transmission

```python
transmission=channel.transmission
```

Define the wavelength 

```python
wavelength=channel.wavelength 
```

```python
def plot_transmission():
    import matplotlib.pyplot as plt
    plt.figure(figsize=(18, 10))

    plt.plot(wavelength,transmission,label=f'{channel.name}')
    
    plt.title('X-Ray Telescope',fontsize = 30)
    plt.xlabel(r'$\lambda$ [Ã…]',fontsize = 30)
    plt.ylabel(r'Transmittance',fontsize=30)
    plt.legend(fontsize=30)
    plt.xlim(-5,80)
    plt.xticks(fontsize=27)
    plt.yticks(fontsize=27)


    plt.grid(color='lightgrey')
    plt.show()  
```

Run plot transmission function

```python
plot_transmission()
```
